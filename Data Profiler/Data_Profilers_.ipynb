<h2 style="color:#27AE60;">📊 Univariate Charts</h2>

<p style="background-color:#27AE60; padding:10px; border-radius:8px;">
<b>Definition:</b><br>
Visually displaying the distribution or frequency of a single variable — using histogram, boxplot, countplot, pie chart.
</p>

<h3 style="color:#27AE60;">📌 Charts Used</h3>
<p>Histplot (distribution), Boxplot (spread/outliers), Countplot & Pie Chart (categorical frequency)</p>

<h3 style="color:#27AE60;">🎯 Purpose</h3>
<p>To understand the shape, outliers, and category distribution of data.</p>

```python
fig , axes = plt.subplots(1,3, figsize = (15,6))

sns.set_style("whitegrid") 
# White background

sns.histplot(df["median_income"],
             kde=True,
             color="#7A9E7E",
             bins=30 , 
             ax=axes[0])
axes[0].set_title(" Distribution of Median Income", fontsize=14, weight = "bold")
axes[0].set_xlabel("Median Income", weight="bold")
axes[0].set_ylabel("Frequency", weight="bold")

sns.boxplot(x=df["median_house_value"]
            ,color = "#D6CFC7", 
            ax=axes[1])
axes[1].set_title(" House Value Spread (Outliers)", fontsize=14, weight="bold")
axes[1].set_xlabel("House Value", weight="bold")

sns.countplot(x=df["ocean_proximity"],
              palette="Greens"
              ,ax=axes[2])
axes[2].set_title(" Ocean Proximity Distribution", fontsize=14, weight="bold")
axes[2].set_xlabel("Ocean Proximity", weight="bold")
axes[0].set_ylabel("Count", weight="bold")

plt.tight_layout()

plt.show()
```

```python
fig , axes = plt.subplots(1,3, figsize = (18,5))

sns.set_style("whitegrid") 
# White background

sns.countplot(x=df["ocean_proximity"],
              palette="Blues",
              ax=axes[0])
axes[0].set_title("Count Plot", weight="bold")
axes[0].set_xlabel("Ocean Proximity")
axes[0].set_ylabel("Count")


df["ocean_proximity"].value_counts().plot(
    kind="bar",
    color="#7A9E7E",
    ax=axes[1]
)
axes[1].set_title("Bar Chart", weight="bold")
axes[1].set_xlabel("Ocean Proximity")
axes[1].set_ylabel("Count")

df['ocean_proximity'].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    color="Greens",
    ax=axes[2]
)
axes[2].set_title("Pie Chart", weight="bold")


plt.tight_layout()

plt.show()
```

<h1 style="color:#E0E0E0; text-align:center;">📘 Bivariate Analysis</h1>

<div style="background-color:#1E1E1E; padding:18px; border-radius:12px; line-height:1.8; font-family:Segoe UI;">

<p style="color:#B0BEC5;">
<b style="color:#64B5F6;">Bivariate Analysis:</b><br>
Analyzing the relationship between two variables. There are three types:<br><br>
<span style="color:#CFD8DC;">
• Numerical vs Numerical → Scatter Plot<br>
• Categorical vs Numerical → Boxplot, Barplot, Violinplot<br>
• Categorical vs Categorical → Heatmap / Cross-tab (if applicable)
</span>
</p>

</div>

```python
plt.figure(figsize=(6,4))

sns.scatterplot(
    x=df["median_income"],
    y=df["median_house_value"],
    color="#6D8B74"
)

plt.title("Income vs House Value", weight="bold")
plt.xlabel("Median Income")
plt.ylabel("Median House Value")
plt.xticks(rotation=30)

plt.show()
```

<h2 style="color:#8E44AD;">🟣 Categorical vs Numerical Analysis</h2>

<p style="background-color:#8E44AD; padding:10px; border-radius:8px;">
<b>Definition:</b><br>
Visualizing the relationship between one categorical and one numerical variable — using boxplot, barplot, violinplot.
</p>

<h3 style="color:#8E44AD;">📌 Example</h3>
<p>ocean_proximity (category) vs median_house_value (number)</p>

<h3 style="color:#8E44AD;">🎯 Purpose</h3>
<p>To see the distribution of numerical values within each category.</p>

```python
plt.figure(figsize=(6,4))

sns.boxplot(
    x=df['ocean_proximity'],
    y=df["median_house_value"],
    palette="Greens"
)

plt.title("Ocean Proximity vs House Value",weight="bold")
plt.xlabel("Ocean Proximity")
plt.ylabel("Median House Value")
plt.xticks(rotation=30)

plt.show()
```

```python
plt.figure(figsize=(18,6))

sns.barplot(
    x=df["ocean_proximity"],
    y=df["median_house_value"],
    palette="Blues"
)

plt.title("Average House Value by Ocean Proximity", weight="bold")
plt.xlabel("Ocean Proximity")
plt.ylabel("Average Value")
plt.xticks(rotation=30)

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.violinplot(
    x=df["ocean_proximity"],
    y=df["median_house_value"],
    data=df,
    palette="Greens"
)

plt.title("Distribution of House Value by Ocean Proximity", weight="bold")
plt.xlabel("Ocean Proximity")
plt.ylabel("Median House Value")

plt.xticks(rotation=30)

plt.show()
```

<h1 style="color:#E0E0E0; text-align:center;">📘 Multivariate Analysis</h1>

<div style="background-color:#1E1E1E; padding:18px; border-radius:12px; line-height:1.8; font-family:Segoe UI;">

<p style="color:#B0BEC5;">
<b style="color:#CE93D8;">Multivariate Analysis:</b><br>
Analyzing three or more variables together at once. This reveals complex patterns and interactions that are not visible when looking at variables individually.
</p>

<p style="color:#B0BEC5;">
<b style="color:#FFCC80;">Correlation Heatmap:</b> Shows the correlation strength between numerical variables.<br>
<b style="color:#A5D6A7;">Pairplot:</b> Displays multiple variable pairs together in scatter plots.<br>
<b style="color:#81D4FA;">Multivariate Scatter:</b> Encodes a 3rd/4th variable using size and hue.<br>
<b style="color:#EF9A9A;">FacetGrid:</b> Repeats the same chart across different categories.
</p>

</div>

```python
plt.figure(figsize=(8,6))

corr = df.corr(numeric_only=True)

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    linewidths=0.5
)

plt.title("Correlation Heatmap", weight="bold")

plt.show()
```

```python
sns.pairplot(
    df[["median_income","median_house_value","housing_median_age"]],
    hue="housing_median_age"
    # extra variable
)

plt.xticks(rotation=30)

plt.show()
```

```python
plt.figure(figsize=(7,5))

sns.scatterplot(
    x=df["median_income"],
    y=df["median_house_value"],
    hue=df["ocean_proximity"],
    # 3rd variable
    size=df["total_rooms"],
    # size change of dot
    sizes=(20,200)
    # dot size
)

plt.title("Multivariate Scatter Plot", weight="bold")
plt.xlabel("Median Income")
plt.ylabel("House Value")

plt.show()
```

```python
palette = ["red", "blue", "green", "purple", "orange"]

g = sns.FacetGrid(
    df,
    col="ocean_proximity",
    col_wrap=3,
    # 3 plots per row
    hue="ocean_proximity",
    palette=palette
)

g.map(sns.histplot, "median_house_value")

g.add_legend()
plt.show()
```
