## 4. 🧮 Tensors

### 📌 Definition
A Tensor is a multi-dimensional array used to store numerical data.
Scalars, vectors, and matrices are all types of tensors.

### 📊 Example

```python
import numpy as np

scalar = np.array(5)           # 0D tensor
vector = np.array([1, 2, 3])  # 1D tensor
matrix = np.array([[1, 2],    # 2D tensor
                   [3, 4]])

print("Scalar:", scalar)
print("Vector:", vector)
print("Matrix:\n", matrix)
```

### 🎯 Purpose
- Used in Deep Learning and Neural Networks
- Efficient data storage
- Faster mathematical computations

## 8. 📄 JSON Parsing

### 📌 Definition
JSON Parsing is the process of extracting data from a JSON file format.

## 9. 🗄️ SQL Data Fetching

### 📌 Definition
Fetching data from a database using SQL queries.

### 📊 Example

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("db.db")
df = pd.read_sql_query("SELECT * FROM customers", conn)
print(df.head())
```

### 🎯 Purpose
- To access large datasets
- To perform real-time data analysis
